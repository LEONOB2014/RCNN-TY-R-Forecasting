```python
%matplotlib inline
#import useful modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import datetime as dt
from mpl_toolkits.basemap import Basemap

#import torch modules
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision import transforms, utils

# import TyDataset and ToTensor class
from dataload_30X30 import TyDataset, ToTensor
```

```python
ty_Train_dataset = TyDataset(excel_file="../颱風事件表.xlsx",
                             train=True,
                             root_dir="../01_TY_database/03_wrangled_data/",
                             transform = ToTensor())
ty_Test_dataset = TyDataset(excel_file="../颱風事件表.xlsx",
                            train=False,
                            root_dir="../01_TY_database/03_wrangled_data/",
                            transform = ToTensor())

trainloader = DataLoader(ty_Train_dataset, batch_size=4,shuffle=True,num_workers=4)
testloader = DataLoader(ty_Test_dataset, batch_size=4,shuffle=True)
```

```python
# Set CNN regression model
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(6, 10, 6)
        self.conv2 = nn.Conv2d(10, 14, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv3 = nn.Conv2d(14, 20, 3)
        self.fc1 = nn.Linear(20 * 30 * 30, 6000)
        self.fc2 = nn.Linear(6000, 3000)
        self.fc3 = nn.Linear(3000, 900)
        
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool(F.relu(self.conv2(x)))
        x = F.relu(self.conv3(x))
        x = x.view(-1, 20 * 30 * 30)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    
net = Net()

import torch.optim as optim
criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.0001, momentum=0.9)
```

```python
for epoch in range(10):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # get the inputs
        inputs, labels = data["RAD"], data["QPE"]
        
        # zero the parameter gradients
        optimizer.zero_grad()
        # forward + backward + optimize
        outputs = net(inputs)
        if labels.shape[0] < 4:
            continue
        else:
            labels = labels.view(4, 900)
        loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer.step()
        print(loss.item())
        # print statistics
#         running_loss += loss.item()
#         if i % 200 == 199:    # print every 2000 mini-batches
#             print('[%d, %5d] loss: %.3f' %
#                   (epoch + 1, i + 1, running_loss / 200))
#             running_loss = 0.0

print('Finished Training')
```

```python
ty_dataset = TyDataset(excel_file="../颱風事件表.xlsx",root_dir="../01_TY_database/03_wrangled_data/")

lat_l=24.55
lat_h=25.45
lon_l=121.15
lon_h=122.05
levels_qp = [-10,0,10,20,30,45,60,80,110]
c_qp = ('#FFFFFF','#E9E9FF','#D2D2FF','#BBBBFF','#AAAAFF','#9090FF','#6060FF','#0000FF')
levels_rad = [-10,0,10,20,30,40,50,60]
c_rad = ('#FFFFFF','#FFE8E8','#FFC8C8','#FFAAAA','#FF9090','#FF6060','#FF0000')

for i in range(len(ty_dataset)):
    if i % 300 == 0:
        plt.figure(figsize=(12,5))
        sample = ty_dataset[i]
        print(i, sample['RAD'].shape, sample['QPE'].shape)
        ax = plt.subplot(2, 4, 1)
        ax.set_title('QPE\n{}'.format(sample['QPE_t']),fontsize=8)
        x = np.linspace(lon_l,lon_h,73)
        y = np.linspace(lat_l,lat_h,73)
        X,Y = np.meshgrid(x,y)
        m = Basemap(projection='cyl',resolution='h', llcrnrlat=lat_l, urcrnrlat=lat_h, llcrnrlon=lon_l, urcrnrlon=lon_h)
        m.readshapefile('../02_TY_raindata/TW_shapefile/gadm36_TWN_2', name='Taiwan', linewidth=0.25, drawbounds=True, color='k')
        cp = m.contourf(X,Y,sample['QPE'],levels_qp,colors=c_qp,alpha=0.95)
        cbar=plt.colorbar(cp)
        cbar.set_label('Rainfall (mm)',fontsize=8)
        cbar.ax.tick_params(labelsize=7)
        plt.xlabel(r"lon$(^o)$",fontsize = 8)
        plt.ylabel(r"lat$(^o)$",fontsize = 8)
        plt.xticks(np.linspace(lon_l,lon_h,5), fontsize = 6)
        plt.yticks(np.linspace(lat_l,lat_h,5), fontsize = 6)

        for j in range(6):
            tmp =1
            if j >= 3:
                tmp = 2
            ax = plt.subplot(2, 4, 1 + j + tmp)
            ax.set_title('RADAR\n{}'.format(sample['RAD_e']-dt.timedelta(minutes=10*j)),fontsize=8)
            m = Basemap(projection='cyl',resolution='h', llcrnrlat=lat_l, urcrnrlat=lat_h, llcrnrlon=lon_l, urcrnrlon=lon_h)
            m.readshapefile('../02_TY_raindata/TW_shapefile/gadm36_TWN_2', name='Taiwan', linewidth=0.25, drawbounds=True, color='k')
            cp = m.contourf(X,Y,sample['RAD'][6-j-1,:,:],levels_rad,colors=c_rad,alpha=0.95)
            cbar=plt.colorbar(cp)
            cbar.set_label('Radar reflection (dbz)',fontsize=8)
            cbar.ax.tick_params(labelsize=7)
            plt.xlabel(r"lon$(^o)$",fontsize = 8)
            plt.ylabel(r"lat$(^o)$",fontsize = 8)
            plt.xticks(np.linspace(lon_l,lon_h,5), fontsize = 6)
            plt.yticks(np.linspace(lat_l,lat_h,5), fontsize = 6)

        plt.tight_layout()
        plt.savefig("sample #{:d}.png".format(int(i/300)),dpi=300,bbox_inches="tight")
        plt.close()
```
